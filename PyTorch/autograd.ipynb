## Example

We manually calculate the derivatives of $f(x) = 3x^2 + 7x$ in $[0, 10]$

```python
import numpy as np
import torch
import matplotlib.pyplot as plt

def fx(x):
    return 3 * x ** 2 + 7 * x

def dfx(x):
    return 6 * x + 7

x = np.linspace(0, 10, 1000)
fx_value = fx(x)
dfx_value = dfx(x)

plt.plot(x, fx_value, label="fx")
plt.plot(x, dfx_value, label="dfx")
plt.legend()
plt.grid(True)
plt.show()
```

We calculate the derivatives of $f(x) = 3x^2 + 7x$ in $[0, 10]$ using autograd in PyTorch

```python
import numpy as np
import torch
import matplotlib.pyplot as plt

def fx(x):
    return 3 * x ** 2 + 7 * x

x = torch.linspace(0, 10, 1000, requires_grad=True)
fx_value = fx(x)

fx_value.sum().backward()

dfx_value = x.grad.numpy()
x = x.detach().numpy()
fx_value = fx_value.detach().numpy()

plt.plot(x, fx_value, label="fx")
plt.plot(x, dfx_value, label="dfx")
plt.legend()
plt.grid(True)
plt.show()
```

## Summary of commom errors in AutoGrad

1. RuntimeError: One of the differentiated Tensors does not require grad
   > Solution: set `require_grad=True`
2. RuntimeError: grad can be implicitly created only for scalar outputs
   > When calculating gradients, the loss must be a scalar. If it is a tensor, the above errors will occur
